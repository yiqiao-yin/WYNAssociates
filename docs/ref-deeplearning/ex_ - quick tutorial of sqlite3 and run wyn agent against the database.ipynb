```python
import sqlite3
```

```python
# Connect to SQLite database
conn = sqlite3.connect('example.db')
c = conn.cursor()
```

```python
# Creating the table if not already exists
c.execute('''CREATE TABLE IF NOT EXISTS stocks
             (date text, trans text, symbol text, qty real, price real)''')
```

```python
# Sample data to insert into the stocks table
stock_data = [
    ('2025-01-05', 'BUY', 'AAPL', 50, 143.77),
    ('2025-01-06', 'SELL', 'GOOGL', 30, 2768.23),
    ('2025-01-07', 'BUY', 'MSFT', 20, 299.02),
    ('2025-01-08', 'SELL', 'TSLA', 15, 1052.32),
    ('2025-01-09', 'BUY', 'AMZN', 10, 3342.88),
    ('2025-01-10', 'SELL', 'FB', 25, 251.64),
    ('2025-01-11', 'BUY', 'NFLX', 40, 489.52),
    ('2025-01-12', 'SELL', 'AAPL', 50, 150.00),
    ('2025-01-13', 'BUY', 'NVDA', 30, 304.68),
    ('2025-01-14', 'SELL', 'BABA', 60, 120.85)
]
```

```python
stock_data
```

```python
# Inserting each row of data
for data in stock_data:
    c.execute("INSERT INTO stocks VALUES (?, ?, ?, ?, ?)", data)
```

```python
# Committing the changes to the database
conn.commit()
```

```python
# Fetching and displaying all rows in the stocks table
for row in c.execute('SELECT * FROM stocks'):
    print(row)
```

```python
# Close the connection to the database
conn.close()
```

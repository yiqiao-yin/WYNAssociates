```python
import requests
import json

def invoke_api(api_url: str, payload: dict, headers: dict) -> dict:
    """
    Sends a POST request to the specified API Gateway endpoint with a given payload and headers.

    Args:
        api_url (str): The URL of the API Gateway endpoint.
        payload (dict): The JSON payload to send in the request.
        headers (dict): The headers to include in the request.

    Returns:
        dict: The parsed response from the API or an error message if the request fails.
    """
    try:
        # Send the request
        response = requests.post(api_url, headers=headers, json=payload)

        # Raise an error for unsuccessful status codes
        response.raise_for_status()

        # Parse and return the JSON response
        return response.json()

    except requests.exceptions.RequestException as e:
        return {"error": "API request failed", "details": str(e)}
```

```python
import base64
from typing import Union

def image_to_base64(file_path: str) -> Union[str, None]:
    """
    Reads an image file from the given file path and returns its content encoded in base64.

    Args:
        file_path (str): The path to the image file (e.g., png, jpg).

    Returns:
        Union[str, None]: The base64 encoded content of the image file, or None if the file could not be read.
    """
    try:
        # Open the image file in binary read mode
        with open(file_path, "rb") as image_file:
            # Read the entire content of the file
            image_content = image_file.read()

            # Encode the content to base64
            base64_encoded_image = base64.b64encode(image_content)

            # Decode the base64 bytes to a string
            base64_string = base64_encoded_image.decode('utf-8')

            return base64_string
    except Exception as e:
        # Handle exceptions (e.g., file not found, permission issues)
        print(f"An error occurred: {e}")
        return None
```

```python
from PIL import Image, ImageDraw, ImageFont
import textwrap

def add_caption_to_img(image_path: str, response_data: dict) -> None:
    """
    Adds a caption extracted from response_data to the bottom of an image and displays it.

    Args:
        image_path (str): The file path of the image.
        response_data (dict): The JSON response containing the caption text.
    """

    try:
        # Load the image
        print(f"Loading image from: {image_path}")
        image = Image.open(image_path)
        print(f"Image loaded successfully. Size: {image.size}")

        # Extract text from response_data
        content_list = response_data.get("model_response", {}).get("content", [])
        text = "\n".join(item.get("text", "") for item in content_list if item.get("type") == "text")

        if not text:
            print("No text found in response data.")
            return

        print(f"Extracted Text: {text}")

        # Define font settings (use a default font)
        try:
            font = ImageFont.truetype("arial.ttf", 20)
        except IOError:
            print("Arial font not found, using default font.")
            font = ImageFont.load_default()

        # Wrap text to fit image width
        max_width = image.width - 20  # Leave padding
        wrapped_text = textwrap.fill(text, width=80)  # Adjust width as needed
        print(f"Wrapped Text: {wrapped_text}")

        # Calculate text size
        draw = ImageDraw.Draw(image)
        text_size = draw.textbbox((0, 0), wrapped_text, font=font)
        text_width, text_height = text_size[2] - text_size[0], text_size[3] - text_size[1]

        print(f"Text Width: {text_width}, Text Height: {text_height}")

        # Create new image with extra space for text
        new_height = image.height + text_height + 20  # Add padding
        print(f"New Image Height: {new_height}")

        new_image = Image.new("RGB", (image.width, new_height), "white")
        print("New image created.")

        # Paste the original image onto the new image
        new_image.paste(image, (0, 0))
        print("Original image pasted onto new image.")

        # Draw text at the bottom
        draw = ImageDraw.Draw(new_image)
        text_position = (10, image.height + 10)
        draw.text(text_position, wrapped_text, fill="black", font=font)
        print(f"Text drawn at position: {text_position}")

        # Display the image
        return new_image
        print("Image displayed successfully.")

    except Exception as e:
        print(f"Error processing image: {str(e)}")
```
